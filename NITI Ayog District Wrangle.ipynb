```python
def extract_names(string):
    nlist = re.findall(r"([a-zA-Z]+)",string)
    state = ''
    district = ''
    for i,v in enumerate(nlist):
        if v == 'UNION':
            continue
        if v == 'STATE' or v == 'TERRITORY':
            flag = False
        elif v != 'DISTRICT' and flag == False:
            state += v 
            if nlist[i+1] != 'DISTRICT':
                state += ' ' 
        elif nlist[i] == 'DISTRICT':
            flag = True
        if flag == True and v != 'DISTRICT':
            district += v
            if i+1 != len(nlist):
                district += ' '
    return [state,district]           
```
