```python
base = 'https://www.niti.gov.in/niti/file/'
i=1
c=0
j=1
while(i is not None):
    url = base + str(i)
    r = requests.get(url,verify=False)
    html_doc = r.text
    soup = BeautifulSoup(html_doc)
    link = soup.find(type=re.compile("application/vnd.openxmlformats-officedocument.spreadsheetml.sheet"))
    if link is None:
        c += 1
        if c>5:
            break
        i += 1    
        continue    
    else:  
        c=0
        print(link.get('href'),i,j)
        j += 1
    i += 1
```
